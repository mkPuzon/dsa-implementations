# Data Structures

## Stack

````python
from implementations.stack import Stack

def test_stack() -> None:
    s = Stack()
    
    # test behavior when empty
    print(f"Peeked: {s.peek()}")
    print(f"Popped: {s.pop()}")
    print(s)
    assert s.is_empty() == True
    
    # test adding items
    s.add(1)
    s.add(2)
    s.add(3)
    assert s.size() == 3
    print("Added items")
    print(s)
    print(f"Popped: {s.pop()}")
    print(f"Peeked: {s.peek()}")
    print(s)
    assert s.size() == 2
    
test_stack()
````

Expected Output:
```
Peeked: None
Popped: None
Stack(size=0)
Added items
Stack(size=3, top=3)
Popped: 3
Peeked: 2
Stack(size=2, top=2)
```

## Queue

````python
from implementations.queue import Queue

def test_queue():
    q = Queue()
    q.peek()
    q.dequeue()
    assert q.is_empty()
    print(q)
    
    q.enqueue('a')
    q.enqueue('b')
    q.enqueue('c')
    assert q.size() == 3

    print(f"Peeked {q.peek()}")
    print(f"Popped {q.dequeue()}")
    print(f"Popped {q.dequeue()}")
    q.enqueue('d')
    assert q.size() == 2
    print(q)

test_queue()
````

Expected Output:

```
Queue(size=0)
Peeked a
Popped a
Popped b
Queue(size=2, next=c)
```
